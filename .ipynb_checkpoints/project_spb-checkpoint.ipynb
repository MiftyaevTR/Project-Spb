### Предобработка данных

```python
# print(data.isna().sum())
data['balcony']=data['balcony'].fillna(0)
data['is_apartment']=data['is_apartment'].fillna(False)
data['locality_name'].dropna(inplace=True)
data['floors_total'].dropna(inplace=True)
data['first_day_exposition']=pd.to_datetime(data['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
#display(data.head())
#data.info()

data['locality_name']=data['locality_name'].replace('деревня Кудрово','Кудрово')
data['locality_name']=data['locality_name'].replace(['поселок городского типа Рябово','посёлок городского типа Рябово'],'поселок Рябово')
data['locality_name']=data['locality_name'].replace('поселок Тельмана','посёлок Тельмана')
data['locality_name']=data['locality_name'].replace(['посёлок Мурино','Мурино'],'поселок Мурино')
data['locality_name']=data['locality_name'].replace(['городской посёлок Рощино','поселок городского типа Рощино'],'Рощино')
data['locality_name']=data['locality_name'].replace('поселок Пансионат Зелёный Бор','посёлок Пансионат Зелёный Бор')
data['locality_name']=data['locality_name'].replace('поселок городского типа Красный Бор','посёлок городского типа Красный Бор')
data['locality_name']=data['locality_name'].replace('поселок Возрождение','посёлок Возрождение')
data['locality_name']=data['locality_name'].replace('поселок городского типа Кузьмоловский','посёлок городского типа Кузьмоловский')
#print(data['locality_name'].unique())
```

* Чаще всего, данные пропущены, так как люди ошибочно не указываю те или иные данный, либо нечего указывать. К примеру с балконами. Если балкона нет, то данные не указываются и остаётся пропуск
* Я изменил тип данный в столбце "first_day_exposition", так как мы не сможем работать с датами, которые имеют тип данных "object".

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Но сами решения по данным и их описание отличные

* Удалим строки с кухнами менее 2кв.м. и больше 50кв.м.

### Посчитайте и добавьте в таблицу новые столбцы

```python
data['price_sq_metr']=data['last_price']/data['living_area']
data['day_of_week'] = data['first_day_exposition'].dt.weekday+1
data['month_of_exposition']=pd.DatetimeIndex(data['first_day_exposition']).month
data['year_of_exposition']=pd.DatetimeIndex(data['first_day_exposition']).year

def categories_floor(row):
    if row['floor']==row['floors_total']:
        return "последний"
    if row['floor']==1:
        return "первый"
    else:
        return "другой"

data['categories_floor'] = data.apply(categories_floor, axis=1)
data['cityCenters_nearest']=data['cityCenters_nearest']/1000
data['cityCenters_nearest']=round(data['cityCenters_nearest'])
data['cityCenters_nearest'].unique()
#display(data.head(10))
```
